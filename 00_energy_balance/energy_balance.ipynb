```python
import pandas as pd
```

```python
data_file = 'Grass_Energy Balance.xlsx'
df1 = pd.read_excel(data_file)
df1 = df1.drop(index=0).reset_index(drop=True)
df1.head()
```

```python
df1['date'] = pd.to_datetime(df1['Date_Time']).dt.date
df1 = df1.drop(columns='Date_Time')
df1.head()
```

```python
df1.isna().sum()
```

```python
df1.info()
```

```python
columns = df1.columns[:-1]
columns
```

```python
for col in columns:
    df1[col] = pd.to_numeric(df1[col])
df1.info()
```

```python
df1 = df1.groupby(df1['date']).mean().reset_index()
```

```python
df1.to_excel('Grass_Energy Balance_Daily.xlsx')
```
