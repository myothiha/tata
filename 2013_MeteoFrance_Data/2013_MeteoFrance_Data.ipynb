```python
import pandas as pd
```

```python
df_raw = pd.read_excel("2013_MeteoFrance_Data.xlsx")
df_raw.head()
```

```python
df_expanded = df_raw.loc[df_raw.index.repeat(4)].reset_index(drop=True)
df_expanded
```

```python
# Add 0, 15, 30, and 45 minutes to each timestamp
df_expanded['DATE'] = df_expanded.groupby(df_expanded.index // 4).cumcount() * 15 * pd.Timedelta(minutes=1) + df_expanded['DATE']
df_expanded
```

```python
df_expanded.to_excel('2013_MeteoFrance_Data_15Min_Data.xlsx')
```
