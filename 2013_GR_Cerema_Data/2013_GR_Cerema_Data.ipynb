```python
import pandas as pd
```

```python
df_raw = pd.read_excel("2013_GR_Cerema_Data.xlsx")
df_raw.head()
```

```python
# Function to convert timestamps to 15-minute intervals
def to_15_min_intervals(ts):
    return ts - pd.to_timedelta(ts.minute % 15, unit='m')

# Apply the function to the 'timestamp' column
df_raw['date'] = df_raw['date'].apply(to_15_min_intervals)
df_raw
```

```python
new_df = df_raw.groupby(df_raw['date']).mean().reset_index()
new_df['Rain'] = df_raw.groupby(df_raw['date'])['Rain'].sum().reset_index()['Rain']
new_df
```

```python
new_df.to_excel("GR_Crema_Every_15min.xlsx")
```
