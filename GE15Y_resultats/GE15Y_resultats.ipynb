```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("GE15Y_resultats.xlsx")
df
```

```python
df.date.dt.month
```

```python
df.Rain
```

```python
df.columns
```

```python
# Using matplotlib

plt.figure(figsize=(10, 6))
# plt.plot(df.date.dt.dayofyear, df.Rain, color='#54ab6f', label='Rain')
plt.plot(df.date.dt.dayofyear, df.Rn_used, color='blue', label='Rn_used')
plt.plot(df.date.dt.dayofyear, df.H, color='red', label='H')
plt.plot(df.date.dt.dayofyear, df.LE, color='black', label='LE') 
plt.plot(df.date.dt.dayofyear, df.G, color='green', label='G')
plt.title('Every 5 Minutes Rain.')  # Add a title
plt.xlabel('Day')  # Add X-axis label
plt.ylabel('Every 5 Minutes Rain')  # Add Y-axis label
plt.grid(True)  # Add gridlines (optional)
# plt.xticks(rotation=45)  # Rotate X-axis labels for better readability (optional)
plt.tight_layout()  # Adjust layout to not cut off labels
plt.legend(loc='upper left')  # Display the legend
plt.show()
```

```python
between_june_to_sept = np.logical_and(df.date > "2013-06-01", df.date <= "2013-09-30")
df2 = df[between_june_to_sept]
df2.head()
```

```python
# Using matplotlib

plt.figure(figsize=(10, 6))
# plt.plot(df.date.dt.dayofyear, df.Rain, color='#54ab6f', label='Rain')
plt.plot(df2.date.dt.dayofyear, df2.Rn_used, color='blue', label='Rn_used')
plt.plot(df2.date.dt.dayofyear, df2.H, color='red', label='H')
plt.plot(df2.date.dt.dayofyear, df2.LE, color='black', label='LE') 
plt.plot(df2.date.dt.dayofyear, df2.G, color='green', label='G')
plt.title('Every 5 Minutes Rain.')  # Add a title
plt.xlabel('Day')  # Add X-axis label
plt.ylabel('Every 5 Minutes Rain')  # Add Y-axis label
plt.grid(True)  # Add gridlines (optional)
# plt.xticks(rotation=45)  # Rotate X-axis labels for better readability (optional)
plt.tight_layout()  # Adjust layout to not cut off labels
plt.legend(loc='upper left')  # Display the legend
plt.show()
```
