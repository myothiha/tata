```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_ch = pd.read_excel("ET_ch.xlsx")
df_ch.head()
```

```python
df_ch.info()
```

```python
df_eb = pd.read_excel("ET_eb.xlsx")
df_eb.head()
```

```python
# Converting the 'time_step' column to datetime format
df_eb['time_step'] = pd.to_datetime(df_eb['time_step'], format='%d/%m/%Y %H:%M:%S')
```

```python
df_eb.info()
```

```python
print(df_ch.shape)
print(df_eb.shape)
```

```python
df_ch.columns
```

```python
df_eb.columns
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_ch['time_step'], df_ch['ET_Ch'], label='ET_Ch', marker='o')
plt.plot(df_eb['time_step'], df_eb['ET_EB'], label='ET_EB', marker='x')

# Formatting the x-axis to show time only
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%H:%M'))

plt.xlabel('Time')
plt.ylabel('ET Values')
plt.title('ET_Ch and ET_EB over Time')
plt.legend()
plt.grid(True)
plt.show()
```
