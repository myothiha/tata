```python
import pandas as pd
```

```python
df_raw = pd.read_excel('Classeur1.xlsx')
columns_to_remove = ['Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22']
df_raw = df_raw.drop(columns=columns_to_remove)
df_raw.columns = df_raw.iloc[0]
df_raw.head()
```

```python
df = df_raw.iloc[3:].reset_index(drop=True)
df.head()
```

```python
# Convert the timestamp column to datetime type
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'])
```

```python
df.head()
```

```python
df.info()
```

```python
def round_up_30min(dt):
    if dt.minute == 0 or dt.minute == 30:
        return dt
    elif dt.minute < 30:
        return dt.replace(minute=30, second=0, microsecond=0)
    else:
        return dt.replace(minute=0, second=0, microsecond=0) + pd.Timedelta(hours=1)
```

```python
# Apply the function to the TIMESTAMP column
df['TIMESTAMP'] = df['TIMESTAMP'].apply(round_up_30min)
df.head()
```

```python
# Group by the new timestamps and compute the average for other columns
df_30min = df.groupby('TIMESTAMP').mean().reset_index()
df_30min.head()
```

```python
df_30min.to_excel("Classeur1_output.xlsx")
```
